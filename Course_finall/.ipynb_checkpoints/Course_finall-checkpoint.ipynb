Функция для визуализации результатов работы модели:

````python
def evaluate_preds(train_true_values, train_pred_values, test_true_values, test_pred_values):
    print("Train R2:\t" + str(round(r2(train_true_values, train_pred_values), 3)))
    print("Test R2:\t" + str(round(r2(test_true_values, test_pred_values), 3)))
    
    plt.figure(figsize=(18,10))
    
    plt.subplot(121)
    sns.scatterplot(x=train_pred_values, y=train_true_values)
    plt.xlabel('Predicted values')
    plt.ylabel('True values')
    plt.title('Train sample prediction')
    
    plt.subplot(122)
    sns.scatterplot(x=test_pred_values, y=test_true_values)
    plt.xlabel('Predicted values')
    plt.ylabel('True values')
    plt.title('Test sample prediction')

    plt.show()
````

Функция для уменьшения объема датасета:

````python
def reduce_mem_usage(df):
    """ iterate through all the columns of a dataframe and modify the data type
        to reduce memory usage.        
    """
    start_mem = df.memory_usage().sum() / 1024**2
    print('Memory usage of dataframe is {:.2f} MB'.format(start_mem))
    
    for col in df.columns:
        col_type = df[col].dtype
        
        if col_type != object:
            c_min = df[col].min()
            c_max = df[col].max()
            if str(col_type)[:3] == 'int':
                if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)
                elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)
                elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)
                elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                    df[col] = df[col].astype(np.int64)  
            else:
                if c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                    df[col] = df[col].astype(np.float32)
                else:
                    df[col] = df[col].astype(np.float64)
        else:
            df[col] = df[col].astype('category')

    end_mem = df.memory_usage().sum() / 1024**2
    print('Memory usage after optimization is: {:.2f} MB'.format(end_mem))
    print('Decreased by {:.1f}%'.format(100 * (start_mem - end_mem) / start_mem))
    
    return df
````

Оформим обработку датасета и генерацию новых фичей в классы

````python
class DataPreprocessing:
    """Подготовка исходных данных"""

    def __init__(self):
        """Параметры класса"""
        self.medians = None
        self.LifeSquare_per_Square_median = None
        self.LifeSquare_per_Square_quantile = None
        
    def fit(self, X):
        """Сохранение статистик"""       
        # Расчет медиан
        self.medians = X.median()
        self.LifeSquare_per_Square_median = (X['LifeSquare'] / X['Square']).median()
        self.LifeSquare_per_Square_quantile = (X['LifeSquare'] / X['Square']).quantile(.05)
        self.SquareMedians = [X.loc[X['Rooms'] == itm + 1, 'Square'].median() for itm in range(5)]
        self.KitchenSquareMedians = [X.loc[X['Rooms'] == itm + 1, 'KitchenSquare'].median() for itm in range(5)]
    
    def transform(self, X):
        """Трансформация данных"""
        
        # Rooms
        X.loc[(X['Rooms'] >= 6) | (X['Rooms'] == 0), 'Rooms'] = self.medians['Rooms']
        
        # Square
        condition = ((X['Square'].isna()) |
                     (X['Square'] < 10) | 
                     (X['Square'] > 150))
        
        for idx in range(5):
            X.loc[condition & (X['Rooms'] == idx + 1), 'Square'] = self.SquareMedians[idx]
        
        # LifeSquare
        condition = ((X['LifeSquare'].isna())
                    | (X['LifeSquare'] / X['Square'] < self.LifeSquare_per_Square_quantile)
                    | (X['LifeSquare'] / X['Square'] > .8))
        
        X.loc[condition, 'LifeSquare'] = self.LifeSquare_per_Square_median * X['Square']
        
        # KitchenSquare
        
        condition = ((X['KitchenSquare'].isna())
                    | (X['KitchenSquare'] > X['Square'])
                    | (X['KitchenSquare'] > X['LifeSquare'])
                    | (X['KitchenSquare'] > 17)
                    | (X['KitchenSquare'] < 3))
        
        for idx in range(5):
            X.loc[condition & (X['Rooms'] == idx + 1), 'KitchenSquare'] = self.KitchenSquareMedians[idx]
        
        # HouseFloor, Floor
        
        X.loc[(X['HouseFloor'] - X['Floor']) < 0 
              | (X['HouseFloor'] > 50), 'HouseFloor'] = self.medians['HouseFloor']
        X.loc[X['HouseFloor'] - X['Floor'] < 0, 'Floor'] = self.medians['Floor']
        
        # HouseYear
        current_year = datetime.now().year
        
        X.loc[X['HouseYear'] > current_year, 'HouseYear'] = self.medians['HouseYear']
        
        # Healthcare_1
        if 'Healthcare_1' in X.columns:
            X.drop('Healthcare_1', axis=1, inplace=True)
        
        X.fillna(self.medians, inplace=True)
        
        return X
````

````python
class FeatureGenetator():
    """Генерация новых фич"""
    
    def __init__(self):
        self.binary_to_numbers = None
        self.MeanPriceDistrict = None
        
    def fit(self, X, y=None):
              
        # Binary features
        self.binary_to_numbers = {'A': 1, 'B': 0}
                
        # Target encoding
        ## District
        df = X.copy()
        
        if y is not None:
            df['Price'] = y.values
            df['Price_1m'] = df['Price'] / df['Square']
            
            self.MeanPriceDistrict = df.groupby('DistrictId', as_index=False).agg({'Price_1m':'mean'})\
                                            .rename(columns={'Price_1m':'MeanPriceDistrict'})
            
            self.MeanPriceDistrict_median = self.MeanPriceDistrict['MeanPriceDistrict'].median()
            
    def transform(self, X):
        
        # Binary features
        X['Ecology_2'] = X['Ecology_2'].map(self.binary_to_numbers)  # self.binary_to_numbers = {'A': 1, 'B': 0}
        X['Ecology_3'] = X['Ecology_3'].map(self.binary_to_numbers)
        X['Shops_2'] = X['Shops_2'].map(self.binary_to_numbers)
        
        # DistrictId
        if self.MeanPriceDistrict is not None:
            X = X.merge(self.MeanPriceDistrict, on='DistrictId', how="left")
            X.fillna(self.MeanPriceDistrict_median, inplace=True)
        
        # LifeSquare_per_Square
        X['LifeSquare_per_Square'] = X['LifeSquare'] / X['Square']
        
        # Приведение типов и уменьшение объема датасета
        X['HouseFloor'] = X['HouseFloor'].astype(np.int64)
        X['Rooms'] = X['Rooms'].astype(np.int64)
        X = self.reduce_mem_usage(X)
        
        return X
    
    @staticmethod
    def reduce_mem_usage(df):
        """ iterate through all the columns of a dataframe and modify the data type
            to reduce memory usage.        
        """
            
        for col in df.columns:
            col_type = df[col].dtype
        
            if col_type != object:
                c_min = df[col].min()
                c_max = df[col].max()
                if str(col_type)[:3] == 'int':
                    if c_min > np.iinfo(np.int8).min and c_max < np.iinfo(np.int8).max:
                        df[col] = df[col].astype(np.int8)
                    elif c_min > np.iinfo(np.int16).min and c_max < np.iinfo(np.int16).max:
                        df[col] = df[col].astype(np.int16)
                    elif c_min > np.iinfo(np.int32).min and c_max < np.iinfo(np.int32).max:
                        df[col] = df[col].astype(np.int32)
                    elif c_min > np.iinfo(np.int64).min and c_max < np.iinfo(np.int64).max:
                        df[col] = df[col].astype(np.int64)  
                else:
                    if c_min > np.finfo(np.float32).min and c_max < np.finfo(np.float32).max:
                        df[col] = df[col].astype(np.float32)
                    else:
                        df[col] = df[col].astype(np.float64)
            else:
                df[col] = df[col].astype('category')
 
        return df
````
