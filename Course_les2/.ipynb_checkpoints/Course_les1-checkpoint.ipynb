```python
import pandas as pd
import numpy as np
```

Прочитаем файл train.csv и посмотрим первые 10 строчек, а также информацию по столбцам

```python
train = pd.read_csv('train.csv')
train.head(10)
```

```python
train.info()
```

Постмотрим статистику по столбцам и начнем их анализировать

```python
train.describe()
```

Обратим внимание на столбец HouseYear. Посмотрим стоки, в которых год постройки дома больше текущего

```python
train.loc[train['HouseYear'] > 2021, 'HouseYear']
```

Как видим, это явные выбросы. Удалим эти строки и посмотрим на статистику столбца без них

```python
train = train.loc[train['HouseYear'] < 2022]
```

```python
train['HouseYear'].describe().apply("{0:.2f}".format)
```

Теперь видно, что в данной строке адекватные значения.

Посмотрим строки, в которых жилая площадь или площадь кухни больше общей площади квартиры, а затем удалим их

```python
train.loc[(train['LifeSquare'] > train['Square']) | (train['KitchenSquare'] > train['Square']), 
          ['Square', 'LifeSquare', 'KitchenSquare']]
```

```python
train = train.loc[~((train['LifeSquare'] > train['Square']) | (train['KitchenSquare'] > train['Square']))]
```

Посмотрим статистику по площадям, а также соотношение жилой к общей площади. Выберем строки, где общая площадь меньше 10 или соотношение жилой площади к общей не входит в диапазон трех сигм с левой стороны, и удалим их. Площадь кухни трогать не будем, т.к. маленькая площадь кухни - это может быть квартира со свободной планировкой или студия совмещенная с кухней

```python
train[['Square', 'LifeSquare', 'KitchenSquare']].describe()
```

```python
tmp = train['LifeSquare'] / train['Square']
tmp.describe()
```

```python
train.loc[(train['Square'] < 10) | (train['LifeSquare'] / train['Square'] < tmp.median() - 3 * tmp.std()),
         ['Square', 'LifeSquare']].describe()
```

```python
train = train.loc[~((train['Square'] < 10) | (train['LifeSquare'] / train['Square'] < tmp.median() - 3 * tmp.std()))]
```

Посмотрим и удалим строки, в которых этажность дома равна 0

```python
train.loc[train['HouseFloor'] == 0, ['HouseFloor']]
```

```python
train = train.loc[~(train['HouseFloor'] == 0)]
```

Теперь посмотрим и удалим те строки, в которых этаж квартиры выше этажности дома

```python
train.loc[train['Floor'] > train['HouseFloor'], ['Floor', 'HouseFloor']]
```

```python
train = train.loc[~(train['Floor'] > train['HouseFloor'])]
```

```python
train.describe()
```

Для столбцов Ecology, Social, Helthcare, Shops  мы не знаем, как рассчитываются эти признаки, чтобы оценить адекватность этих значений. Поэтому просто отфильтруем по правилу трех сигм по верхней границе

```python
train.loc[~((train['Ecology_1'] > train['Ecology_1'].median() + 3 * train['Ecology_1'].std()) | 
         (train['Social_1'] > train['Social_1'].median() + 3 * train['Social_1'].std()) |
         (train['Social_2'] > train['Social_2'].median() + 3 * train['Social_2'].std()) |
         (train['Social_3'] > train['Social_3'].median() + 3 * train['Social_3'].std()) |
         (train['Healthcare_1'] > train['Healthcare_1'].median() + 3 * train['Healthcare_1'].std()) |
        (train['Helthcare_2'] > train['Helthcare_2'].median() + 3 * train['Helthcare_2'].std()) |
         (train['Shops_1'] > train['Shops_1'].median() + 3 * train['Shops_1'].std()))].describe()
```

Как видно, кроме Social_3, в остальных столбцах мало что поменялось в статистике. Посмотрим отдельно этот столбец

```python
train.loc[(train['Social_3'] > train['Social_3'].median() + 3 * train['Social_3'].std()), ['Social_3']].describe()
```

```python
train.loc[(train['Social_3'] == 141)].describe()
```

Значение 141 не выглядит выбросом, потому как коррелируется с другими индексами. Поэтому пока эти столбцы трогать не будем
