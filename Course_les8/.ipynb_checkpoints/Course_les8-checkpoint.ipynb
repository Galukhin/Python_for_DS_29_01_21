```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.image as img
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from scipy.stats import mode

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, KFold, GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error as mse, r2_score as r2
from sklearn.ensemble import RandomForestRegressor

import warnings
warnings.filterwarnings('ignore')
```

```python
DATASET_PATH = './train.csv'
PREPARED_DATASET_PATH = './train_prepared.csv'
```

```python
df = pd.read_csv(PREPARED_DATASET_PATH, sep=';')
df.info()
```

```python
target = 'Price'
y = df[target]
X = df.drop(target, axis=1)
```

```python
scaler = StandardScaler()

cols = X.columns
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=cols)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25, random_state=100)
```

```python
def evaluate_preds(true_values, pred_values, save=False):
    """Оценка качества модели и график preds vs true"""
    
    print("R2:\t" + str(round(r2(true_values, pred_values), 3)) + "\n" +
          "RMSE:\t" + str(round(np.sqrt(mse(true_values, pred_values)), 3)) + "\n" +
          "MSE:\t" + str(round(mse(true_values, pred_values), 3))
         )
    
    plt.figure(figsize=(8,8))
    
    sns.scatterplot(x=pred_values, y=true_values)
    plt.plot([0, 650000], [0, 650000], linestyle='--', color='black')  # диагональ, где true_values = pred_values
    
    plt.xlabel('Predicted values')
    plt.ylabel('True values')
    plt.title('True vs Predicted values')
    
    if save == True:
        plt.savefig(REPORTS_FILE_PATH + 'report.png')
    plt.show()
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gb_model = GradientBoostingRegressor(criterion='mse',
                          min_samples_leaf=10,
                          random_state=65,
                         # n_estimators=100,
                         # max_depth=7
                         )
```

```python
%%time
params = {'n_estimators':[50, 100, 200, 400], 
          'max_depth':np.arange(3, 12)}

gs = GridSearchCV(gb_model, params, 
                  scoring='r2', # метрика 
                  cv=KFold(n_splits=5,   # k (кол-во разбиений/итераций) в кросс-валидации
                           random_state=21, 
                           shuffle=True),
                  n_jobs=-1
                  )
gs.fit(X_train, y_train)
```

```python
gs.best_params_
```

```python
gs.best_score_
```

```python
gb_model = GradientBoostingRegressor(criterion='mse',
                          min_samples_leaf=10,
                          random_state=65,
                          n_estimators=200,
                          max_depth=3
                         )
```

```python
gb_model.fit(X_train, y_train)
```

```python
y_train_preds = gb_model.predict(X_train)
evaluate_preds(y_train, y_train_preds)
```

```python
y_test_preds = gb_model.predict(X_test)
evaluate_preds(y_test, y_test_preds)
```

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2, learning_rate=250, random_state=42)

X_train_tsne = tsne.fit_transform(X_train)
```

```python
plt.scatter(X_train_tsne[:, 0], X_train_tsne[:, 1])

plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
def best_clasters(X_train, y_train, X_test, y_test):
    r2_list = []
    for i in np.arange(2, 16):
        kmeans = KMeans(n_clusters=i, max_iter=100, random_state=42)
        labels_train = kmeans.fit_predict(X_train)
        X_train['claster'] = labels_train
        labels_test = kmeans.predict(X_test)
        X_test['claster'] = labels_test
        gb_model.fit(X_train, y_train)
        y_test_preds = gb_model.predict(X_test)
        r2_list.append(r2(y_test, y_test_preds))
    r2_max = max(r2_list)
    best_cluster = r2_list.index(r2_max) + 2
    print("R2:\t" + str(round(r2_max, 3)) + "\n" +
          "Best quantity of clasters:\t" + str(best_cluster)
         )
```

```python
best_clasters(X_train, y_train, X_test, y_test)
```
